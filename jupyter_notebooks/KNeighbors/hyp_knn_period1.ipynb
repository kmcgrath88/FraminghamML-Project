```python
#import dependencies
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
#read in data set
df = pd.read_csv('../Resources/HYP_final_data.csv')
df.head(20)
```

```python
HYP_df = df.set_index('RANDID')
#Remove columns not involved in analysis
#removed 'TIME', 'PERIOD', 'DEATH', 'TIMEHYP'

HYP_df = HYP_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
       'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVMI', 'PREVCHD',
       'PREVAP', 'PREVSTRK', 'PREVHYP', 'HYPERTEN']]
HYP_df.head()
```

```python
y = HYP_df['HYPERTEN']
target_names = ['negative', 'positive']
```

```python
X = HYP_df.drop('HYPERTEN', axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=49)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f'k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}')
```

```python
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker='x')
plt.xlabel('k neighbors')
plt.ylabel('Testing Accuracy Score')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train_scaled, y_train)
print('k=11 Test Acc: %.3f' % knn.score(X_test_scaled, y_test))
```

```python
#Make predictions (0 = no event occured, 1 = event occurred)
predictions = knn.predict(X_test_scaled)
hyp_predictions = pd.DataFrame({'Prediction': predictions, 'Actual': y_test}).reset_index(drop=True)
hyp_predictions
```

```python
p0a0 = hyp_predictions.loc[(hyp_predictions['Prediction']==0) & (hyp_predictions['Actual']==0)]
p1a1 = hyp_predictions.loc[(hyp_predictions['Prediction']==1) & (hyp_predictions['Actual']==1)]
p0a1 = hyp_predictions.loc[(hyp_predictions['Prediction']==0) & (hyp_predictions['Actual']==1)]
p1a0 = hyp_predictions.loc[(hyp_predictions['Prediction']==1) & (hyp_predictions['Actual']==0)]

print(f"Event prediction = 0 and actual = 0:  {len(p0a0)}")
print(f"Event prediction = 1 and actual = 1:  {len(p1a1)}")
print(f"Event prediction = 0 and actual = 1:  {len(p0a1)}")
print(f"Event prediction = 1 and actual = 0:  {len(p1a0)}")
```

```python
#filter df by gender (males = 1, females = 2)
male_HYP = HYP_df.loc[(HYP_df['SEX']==1)]
female_HYP = HYP_df.loc[(HYP_df['SEX']==2)]
```

```python
male_HYP_df = male_HYP.drop('SEX', axis=1)
female_HYP_df = female_HYP.drop('SEX', axis=1)
```

```python
y_male = male_HYP_df['HYPERTEN']
male_target_names = ['negative', 'positive']

y_female = female_HYP_df['HYPERTEN']
female_target_names = ['negative', 'positive']
```

```python
X_male = male_HYP_df.drop('HYPERTEN', axis=1)
X_female = female_HYP_df.drop('HYPERTEN', axis=1)
```

```python
X_male_train, X_male_test, y_male_train, y_male_test = train_test_split(X_male, y_male, random_state=49)
X_female_train, X_female_test, y_female_train, y_female_test = train_test_split(X_female, y_female, random_state=49)
```

```python
# Create a StandardScater model and fit it to the training data - male

X_male_scaler = StandardScaler().fit(X_male_train)

#Transform the training and testing data using the X_scaler and y_scaler models

X_male_train_scaled = X_male_scaler.transform(X_male_train)
X_male_test_scaled = X_male_scaler.transform(X_male_test)
```

```python
m_train_scores = []
m_test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_male_train_scaled, y_male_train)
    train_score = knn.score(X_male_train_scaled, y_male_train)
    test_score = knn.score(X_male_test_scaled, y_male_test)
    m_train_scores.append(train_score)
    m_test_scores.append(test_score)
    print(f'k: {k}, Train/Test Score: {train_score: .3f}/{test_score:.3f}')
```

```python
plt.plot(range(1, 20, 2), m_train_scores, marker='o')
plt.plot(range(1, 20, 2), m_test_scores, marker='x')
plt.xlabel('k neighbors')
plt.ylabel('Testing Accuracy Score')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_male_train_scaled, y_male_train)
print('k=13 Test Acc: %.3f' % knn.score(X_male_test_scaled, y_male_test))
```

```python
m_predictions = knn.predict(X_male_test_scaled)
m_hyp_predictions = pd.DataFrame({'Prediction': m_predictions, 'Actual': y_male_test}).reset_index(drop=True)
m_hyp_predictions
```

```python
p0a0 = m_hyp_predictions.loc[(m_hyp_predictions['Prediction']==0) & (m_hyp_predictions['Actual']==0)]
p1a1 = m_hyp_predictions.loc[(m_hyp_predictions['Prediction']==1) & (m_hyp_predictions['Actual']==1)]
p0a1 = m_hyp_predictions.loc[(m_hyp_predictions['Prediction']==0) & (m_hyp_predictions['Actual']==1)]
p1a0 = m_hyp_predictions.loc[(m_hyp_predictions['Prediction']==1) & (m_hyp_predictions['Actual']==0)]

print(f"Event prediction = 0 and actual = 0 (male):  {len(p0a0)}")
print(f"Event prediction = 1 and actual = 1 (male):  {len(p1a1)}")
print(f"Event prediction = 0 and actual = 1 (male):  {len(p0a1)}")
print(f"Event prediction = 1 and actual = 0 (male):  {len(p1a0)}")
```
