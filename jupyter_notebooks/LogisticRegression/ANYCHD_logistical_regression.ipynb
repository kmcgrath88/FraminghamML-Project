```python
#import dependencies
import pandas as pd
```

```python
#read in data set
df = pd.read_csv("ANYCHD_final_data.csv")
df.head(20)
```

```python

ANYCHD_df = df.set_index('RANDID')
ANYCHD_df.head()
```

```python

# Number of hospitalizations for ANYCHD infarctions and number of non-events
ANYCHD_count=ANYCHD_df.loc[(ANYCHD_df['ANYCHD']==1)]
no_event_count = ANYCHD_df.loc[(ANYCHD_df['ANYCHD']==0)]

print(f"Subjects hospitalized with ANYCHD: {len(ANYCHD_count)}")
print(f"Subjects with no ANYCHD events: {len(no_event_count)}")
```

```python
ANYCHD_df.columns
```

```python
ANYCHD_reg_df = ANYCHD_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
       'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVCHD', 'PREVAP',
       'PREVMI', 'PREVSTRK', 'PREVHYP', 'HOSPMI', 'ANYCHD', ]]
ANYCHD_reg_df.head()
```

```python
# Assign X (data) and y (target)
X = ANYCHD_reg_df.drop('ANYCHD', axis=1)
y= ANYCHD_reg_df['ANYCHD']
print(X.shape, y.shape)
```

```python

# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
# Fit the model to the data
classifier.fit(X_train, y_train)
```

```python
#Validate the model using test data
# Print the r2 score for the test data
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
