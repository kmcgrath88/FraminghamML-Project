```python
import pandas as pd
```

```python
df = pd.read_csv('../Resources/HYP_final_data.csv')
df.head()
```

```python
HYP_df = df.set_index('RANDID')
HYP_df.head()
```

```python
#Number of events for Hypertension calls and number of non-events
HYP_count = HYP_df.loc[(HYP_df['HYPERTEN']==1)]
no_event_count = HYP_df.loc[(HYP_df['HYPERTEN']==0)]

print(f'Subjects afflcited with HYP: {len(HYP_count)}')
print(f'Subjects with no HYP events: {len(no_event_count)}')
```

```python
HYP_df.columns
```

```python
HYP_reg_df = HYP_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
                    'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVCHD', 'PREVAP', 
                    'PREVMI', 'PREVSTRK', 'PREVHYP', 'HYPERTEN']]
HYP_reg_df.head()
```

```python
X = HYP_reg_df.drop('HYPERTEN', axis=1)
y = HYP_reg_df['HYPERTEN']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f'Training Data Score: {classifier.score(X_train, y_train)}')
print(f'Testing Data Score: {classifier.score(X_test, y_test)}')
```

```python
predictions = classifier.predict(X_test)
HYP_predictions = pd.DataFrame({'Prediction': predictions, 'Actual': y_test}).reset_index(drop=True)
HYP_predictions
```

```python
no_event = HYP_predictions.loc[(HYP_predictions['Prediction']==0) & (HYP_predictions['Actual']==0)]
event = HYP_predictions.loc[(HYP_predictions['Prediction']==1) & (HYP_predictions['Actual']==1)]
p0a1 = HYP_predictions.loc[(HYP_predictions['Prediction']==1) & (HYP_predictions['Actual']==1)]
p1a0 = HYP_predictions.loc[(HYP_predictions['Prediction']==1) & (HYP_predictions['Actual']==0)]
```

```python
print(f"Event prediction and actual = 0:  {len(no_event)}")
print(f"Event prediction and actual = 1:  {len(event)}")
print(f"Event prediction = 0 and actual = 1:  {len(p0a1)}")
print(f"Event prediction = 1 and actual = 0:  {len(p1a0)}")
```
