```python
import pandas as pd
```

```python
df = pd.read_csv('../Resources/ANG_final_data.csv')
df.head()
```

```python
ANG_df = df.set_index('RANDID')
ANG_df.head()
```

```python
# Number of hospitalizations for myocardial infarctions and number of non-events
ANG_count=ANG_df.loc[(ANG_df['ANGINA']==1)]
no_event_count = ANG_df.loc[(ANG_df['ANGINA']==0)]

print(f"Subjects afflicted with Angina: {len(ANG_count)}")
print(f"Subjects with no Angina events: {len(no_event_count)}")
```

```python
ANG_df.columns
```

```python
ANG_reg_df = ANG_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
       'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVCHD', 'PREVAP',
       'PREVMI', 'PREVSTRK', 'PREVHYP', 'ANGINA']]
ANG_reg_df.head()
```

```python
X = ANG_reg_df.drop('ANGINA', axis=1)
y = ANG_reg_df['ANGINA']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f'Training Data Score: {classifier.score(X_train, y_train)}')
print(f'Testing Data Score: {classifier.score(X_test, y_test)}')
```

```python
predictions = classifier.predict(X_test)
ANG_predictions = pd.DataFrame({'Prediction': predictions, 'Actual': y_test}).reset_index(drop=True)
ANG_predictions
```

```python
no_event = ANG_predictions.loc[(ANG_predictions['Prediction']==0) & (ANG_predictions['Actual']==0)]
event = ANG_predictions.loc[(ANG_predictions['Prediction']==1) & (ANG_predictions['Actual']==1)]
p0a1 = ANG_predictions.loc[(ANG_predictions['Prediction']==1) & (ANG_predictions['Actual']==1)]
p1a0 = ANG_predictions.loc[(ANG_predictions['Prediction']==1) & (ANG_predictions['Actual']==0)]
```

```python
print(f"Event prediction and actual = 0:  {len(no_event)}")
print(f"Event prediction and actual = 1:  {len(event)}")
print(f"Event prediction = 0 and actual = 1:  {len(p0a1)}")
print(f"Event prediction = 1 and actual = 0:  {len(p1a0)}")
```
