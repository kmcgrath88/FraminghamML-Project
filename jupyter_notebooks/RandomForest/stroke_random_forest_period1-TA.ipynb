# stroke - Random Forest

```python
from sklearn import tree
import pandas as pd
```

```python
#read in data set
df = pd.read_csv('../../Resources/STROKE_final_data.csv')
df.head()
```

```python
df.columns
```

```python
stroke_df = df.set_index('RANDID')
#Remove columns not involved in analysis
#removed 'TIME', 'PERIOD', 'DEATH', 'TIMEHYP'

stroke_df = stroke_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
       'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVCHD', 'PREVAP',
       'PREVMI', 'PREVHYP', 'STROKE']]
stroke_df.head()
```

```python
target = stroke_df["STROKE"]
target_names = ["negative", "positive"]
```

```python
data = stroke_df.drop("STROKE",axis=1)
feature_names = data.columns
data.head()
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=49)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Create, fit, and score a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=500, class_weight = "balanced")
rf = rf.fit(X_train, y_train)

print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```

```python
# calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# sort the features by their importance
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
#Make predictions (0 = no event occurred, 1 = event occurred)
predictions = rf.predict(X_test)
stroke_predictions = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
stroke_predictions
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
