# Myocardial Infarction - Random Forest

```python
from sklearn import tree
import pandas as pd
```

```python
#read in data set
df = pd.read_csv("../../Resources/MI_final_data.csv")
df.head(20)
```

```python
MI_df = df.set_index('RANDID')
#Remove columns not involved in analysis
#removed 'Time', 'Period', 'Death', 'MI_FCHD', 'TIMEMI', 'TIMEMIFC'
MI_df = MI_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
       'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVCHD', 'PREVAP',
       'PREVMI', 'PREVSTRK', 'PREVHYP', 'HOSPMI']]
MI_df.head()
```

```python
target = MI_df["HOSPMI"]
target_names = ["No_MI", "MI"]
```

```python
data = MI_df.drop("HOSPMI",axis=1)
feature_names = data.columns
data.head()
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=49)
```

```python
# Create a Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_train, y_train)
```

```python
# Fit the classifier to the data & print out the score of the model
print(f"Training Data Score: {clf.score(X_train, y_train)}")
print(f"Testing Data Score: {clf.score(X_test, y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
# Create, fit, and score a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=500, class_weight = "balanced")
rf = rf.fit(X_train, y_train)

print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```

```python
# calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# sort the features by their importance
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
#Make predictions (0 = no event occurred, 1 = event occurred)
predictions = rf.predict(X_test)
MI_predictions = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
MI_predictions
```

```python
no_event = MI_predictions.loc[(MI_predictions['Prediction']==0) & (MI_predictions['Actual']==0)]
event = MI_predictions.loc[(MI_predictions['Prediction']==1) & (MI_predictions['Actual']==1)]
p0a1 = MI_predictions.loc[(MI_predictions['Prediction']==0) & (MI_predictions['Actual']==1)]
p1a0 = MI_predictions.loc[(MI_predictions['Prediction']==1) & (MI_predictions['Actual']==0)]
```

```python
print(f"Event prediction and actual = 0:  {len(no_event)}")
print(f"Event prediction and actual = 1:  {len(event)}")
print(f"Event prediction = 0 and actual = 1:  {len(p0a1)}")
print(f"Event prediction = 1 and actual = 0:  {len(p1a0)}")
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
