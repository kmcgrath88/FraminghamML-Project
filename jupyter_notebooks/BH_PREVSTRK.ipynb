```python
#filter for Period = 1 and PrevMI = 0
PREVSTRK_df = df.loc[(df["PERIOD"]==1) & (df["PREVSTRK"]==0)]
PREVSTRK_df
```

```python
PREVSTRK_df.isnull().sum()
```

```python
PREVSTRK_final_df = PREVSTRK_df.dropna()
PREVSTRK_final_df
```
