ANYCHD_SVM

```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ANYCHD_final_data.csv")
df.head(20)
```

```python
ANYCHD_df = df.set_index('RANDID')
#Remove columns not involved in analysis
#removed 'Time', 'Period', 'Death', 'MI_FCHD', 'TIMEMI', 'TIMEMIFC'
ANYCHD_df = ANYCHD_df[['SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI',
       'DIABETES', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'PREVSTRK', 'PREVHYP', 'ANYCHD']]
ANYCHD_df.head()
```

```python
target = ANYCHD_df["ANYCHD"]
target_names = ["negative", "positive"]
```

```python
data = ANYCHD_df.drop("ANYCHD", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=49)
```

```python

# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
# Adjust `C` with values of 1, 5, and 10. Adjust `gamma` using .0001, 0.001, and 0.01
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python

# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train, y_train)
```

```python
# List the best parameters for this dataset
print(grid.best_params_)
```

```python
# List the best score
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
ANYCHD_predictions = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
ANYCHD_predictions
```
