```python
import pandas as pd
```

```python
df = pd.read_csv('../Resources/frmgham2.csv')
df.head()
```

```python
df.columns
```

```python
#filter for Period = 1 and PREVHYP  = 0
HYP_df = df.loc[(df['PERIOD']==1) & (df['PREVHYP']==0)]
HYP_df
```

```python
#keep the columns we want
HYP_columns = ['RANDID', 'SEX', 'TOTCHOL', 'AGE', 'SYSBP', 'DIABP', 'CURSMOKE', 'CIGPDAY', 'BMI', 'DIABETES', 'BPMEDS',
              'HEARTRTE', 'GLUCOSE', 'PREVCHD', 'PREVAP', 'PREVSTRK', 'PREVHYP', 'TIME', 'PERIOD', 'DEATH', 'HYPERTEN',
              'TIMEHYP']
HYP_df = HYP_df[HYP_columns]
HYP_df
```

```python
#find nulls in our data
HYP_df.isnull().sum()
```

```python
HYP_final_df = HYP_df.dropna()
HYP_final_df
```

```python
HYP_final_df.to_csv('../Resources/HYP_final_data.csv', index=False)
```

```python
HYP_g_df = HYP_df.drop(columns=['GLUCOSE'])
HYP_g_final_df = HYP_g_df.dropna()
HYP_g_final_df
```

```python
HYP_g_final_df.isnull().sum()
```

```python
HYP_g_final_df.to_csv('../Resources/HYP_final_noglucose.csv', index=False)
```
